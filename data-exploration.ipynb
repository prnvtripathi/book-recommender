Checking if the data has some kind of regular inconsistencies

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()

sns.heatmap(books.isna(), ax=ax)

plt.xlabel("Column")
plt.ylabel("Missing Values")

plt.show()
```

Checking if there is any relation between, missing description, num of pages, age of book or average rating

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2025 - books['published_year']
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))

heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
heatmap.set_title("Correlation Matrix", fontdict={"fontsize": 14})

plt.show()
```

Removing books where these are missing: description, num_pages, average rating and published year.

```python
books_missing = books[~(books["description"].isna()) &
                      ~(books["num_pages"].isna()) &
                      ~(books["average_rating"].isna()) &
                      ~(books["published_year"].isna())
                      ]
```

Checking length of the books description, to see if they are useful or not.

```python
books_missing["words_in_description"] = books_missing["description"].str.split().str.len()
```

Creating a new collection with books having description shorter than 25 words

```python
books_missing_25_words = books_missing[books_missing["words_in_description"] >= 25]
```

Joining title and subtitle

```python
books_missing_25_words
```

```python
books_missing_25_words["title_and_subtitle"] = (
    np.where(books_missing_25_words["subtitle"].isna(), books_missing_25_words["title"],
             books_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
books_missing_25_words
```

Adding isbn as the first word of the description

```python
books_missing_25_words["tagged_description"] = books_missing_25_words[["isbn13", "description"]].astype(str).agg(
    " ".join, axis=1)
```

```python
books_missing_25_words
```

Dropping unused columns

```python
(
    books_missing_25_words
    .drop(["subtitle", "missing_description", "words_in_description", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index=False)
)
```
